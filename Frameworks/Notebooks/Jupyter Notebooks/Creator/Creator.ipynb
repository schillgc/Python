```python
def check_sentiment(response):
    for i in range(len(affirmative)):
        affirmative[i] = affirmative[i].lower()
        if response.lower() == affirmative[i]:
            return True
        else:
            for i in range(len(negation)):
                negation[i] = negation[i].lower()
                if response.lower() == negation[i]:
                    return False
```

```python
class Temperature:
    def _init__ (self, temperature = 0):
        self.set_temperature(temperature)

    def fahrenheit (self, temperature):
        return self.temperature

    def celsius (self):
        return (self.get_temperature() - 32) * 5/9
    
    def get_temperature(self):
        return self.temperature
    
    def set_temperature(self, value):
        self.temperature = value
```

```python
def ask_about_air_temperature():
    air = Temperature()

    air.temperature = float(input('What is the air temperature? '))
    should_round = check_sentiment(input('Would you like that rounded? '))

    if should_round:
        print(round(air.fahrenheit()), " F")
        print(round(air.celsius()), " C")
    elif should_round is None:
        print("Please rephrase.  Thanks!")
        ask_about_air_temperature()
    elif not should_round:
        print(air.fahrenheit(), " F")
        print(air.celsius(), " C")
    else:
        print("Sorry an error occurred.")
        ask_about_air_temperature()

ask_about_air_temperature()
```
